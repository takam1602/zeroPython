```python
import matplotlib.pyplot as plt

def logistic(r):
  n=0.1
  for i in range(1000):
    n = r*n*(1.0-n)
    if i>990:
      print(n)
```

```python
logistic(4.0)
```

```python
def logisticPlot(start,end,x,y):
  for i in range(3000):
    r = (end-start)*i/3000 +start
    n = 0.1
    for j in range(3000):
      n = r*n*(1.0-n)
      if j>100:
        x.append(r)
        y.append(n)

def plot(start,end):
  x,y = [],[]
  logisticPlot(start,end,x,y)
  plt.style.use('dark_background')
  plt.figure(figsize=(60,30))
  plt.scatter(x,y,s=0.1,c="green")
```

```python
plot(1.0,3.0)
```

```python
plot(1.0,4.0)
```

```python
def collatzGraph(i,edges):
  while i!=1:
    j=i
    if i%2 ==0:
      i = i//2
    else:
      i =i*3+1
    edges.add((j,i))
    
def makeGraph(n):
  g = Digraph()
  edges = set()
  for i in range(1,n+1):
    collatzGraph(i,edges)
  
  for i,j in edges:
    g.edge(str(i),str(j))
  g.attr(size="10,10")
  return g
```

```python
makeGraph(3)
```

```python
makeGraph(5)
```

```python
makeGraph(19)
```

```python
def collatzExtGraph(i,edges):
  while (i!=1 and i!=3):
    j=i
    if i%2 ==0:
      i = i//2
    else:
      i =i*3+3
    edges.add((j,i))
    
def makeExtGraph(n):
  g = Digraph()
  edges = set()
  for i in range(1,n+1):
    collatzExtGraph(i,edges)
  
  for i,j in edges:
    g.edge(str(i),str(j))
  g.attr(size="10,10")
  return g
```
